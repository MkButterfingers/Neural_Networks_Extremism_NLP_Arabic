```python
import pandas as pd

# Load the CSV files into dataframes
labels_df = pd.read_excel('labels.xlsx',dtype = {'tweet_id':str})
tweets_df = pd.read_excel('batches.xlsx', dtype = {'tweet_id':str})




```

```python
tweets_df
```

```python
# Merge the two dataframes based on the tweet ID
merged_df = pd.merge(tweets_df, labels_df, on='tweet_id')

# Save the merged dataframe to a new CSV file
merged_df.to_csv('merged_tweets.csv', index=False)
```
