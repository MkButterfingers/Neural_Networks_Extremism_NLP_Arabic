```python
!pip install -U snscrape
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
```

```python
my_list = []
for id in tqdm(ids[:2000]):
    scraper = sntwitter.TwitterTweetScraper(id)
    for tweet in scraper.get_items():
        try:
            data = [tweet.id, tweet.content]
        except AttributeError as e:
            print(f"AttributeError: {e} for tweet ID {tweet.id}")
            data = [tweet.id, None]
        my_list.append(data)
```

```python
len(my_list)
```

```python
tweets_df = pd.DataFrame(my_list, columns = ["ids", "text"])
```

```python
tweets_df
```

```python
# Create a set of all the IDs
all_ids = set(ids[:100])

# Create a set of the IDs in my_list
added_ids = set([data[0] for data in my_list])

# Find the IDs that were not added to my_list
not_added_ids = all_ids - added_ids

# Print the IDs that were not added to my_list
print("IDs not added to my_list:")
for id in not_added_ids:
    print(id)
```
